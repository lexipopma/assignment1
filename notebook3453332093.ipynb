```python
# Problem 4

def calculate_grade(percentage):
    if percentage >= 90.0:
        letter = "A"
    elif percentage >= 80.0:
        letter = "B"
    elif percentage >= 70.0:
        letter = "C"
    elif percentage >= 60.0:
        letter = "D"
    else:
        letter = "F"
    return letter
```

```python
print(calculate_grade(90.0))
```

```python
# Problem 5

def calculate_total_cost(num_apples: int, num_bananas: int):

    cost_apples = num_apples * 0.50
    cost_bananas = num_bananas * 0.30
    calculate_total_cost = cost_apples + cost_bananas
    return calculate_total_cost

print(calculate_total_cost(4,5))
```

```python
# Problem 6

def square_and_add(x: float, y: float):
    square_x = x ** 2
    add_y = y
    square_and_add = square_x + add_y
    return square_and_add

print(square_and_add(-2,10))
```

```python
# Problem 7

def greet_by_time(hour : float):
    if 0 < hour < 11:
        greeting = "Good morning!"
    elif 12 < hour < 18:
        greeting = "Good afternoon!"
    elif 19 < hour < 23:
        greeting = "Good evening"
    else:
        greeting = "Invalid time!"
    return greeting

print(greet_by_time(-1))
```
